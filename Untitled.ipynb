```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
# Set up the WebDriver (e.g., for Chrome)
chrome_options = webdriver.ChromeOptions()
chrome_driver_path = "C:/Users/Lenovo/AppData/Local/Temp/Rar$EXa0.922/chromedriver"
# Open the webpage

driver = webdriver.Chrome(options=chrome_options)
# Find the table element (modify the locator according to your website)
driver.get("https://www.nirfindia.org/Rankings/2024/EngineeringRanking.html")
html_content = driver.page_source

wait = WebDriverWait(driver, 10)  # wait up to 10 seconds

# Locate the table (modify the XPath to fit your table structure)
table = wait.until(EC.presence_of_element_located((By.XPATH, "//table")))
buttons = table.find_elements(By.XPATH, "//a[contains(@class, 'btn-more-details')]")

# Iterate through each "More Details" button and click it
for button_idx, button in enumerate(buttons):
    print(f"Clicking 'More Details' button {button_idx + 1}")
    
    # Scroll into view to ensure the button is visible before clicking
    driver.execute_script("arguments[0].scrollIntoView(true);", button)
    
    try:
        # Wait until the button is clickable and click it
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable(button)).click()
        print(f"Clicked button {button_idx + 1}")
        
        # Optionally: Extract additional details that become visible after clicking
        # Example: Assume a new row or section is expanded after clicking the button
        # details_section = driver.find_element(By.XPATH, "xpath_of_expanded_section")
        # print(f"Extracted details: {details_section.text}")
        
    except Exception as e:
        print(f"Error clicking button {button_idx + 1}: {str(e)}")
    time.sleep(1)
# Get all the rows in the table
rows = table.find_elements(By.TAG_NAME, "tr")

# Initialize lists to store the extracted data
table_data = []

# Loop through each row to get the cell data
for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")
    row_data = []
    for cell_idx, cell in enumerate(cells):
        print(f"    Cell {cell_idx}: {cell.text}")
        
    print("-" * 40)  # Just for separating rows visually
    if len(cells) > 0:
        # Extract Institute Code
        institute_code = cells[0].text
        row_data.append(institute_code)
        
        # Extract Institute Name and Links (PDF, Graph, etc.)
        institute_name = cells[1].text.split("\n")[0]  # Extract the name
        row_data.append(institute_name)

        try:
            # Wait for the PDF link to be present and extract it
            pdf_link = WebDriverWait(cells[1], 5).until(EC.presence_of_element_located((By.XPATH, ".//a[contains(@href, '.pdf')]"))).get_attribute("href")
        except:
            pdf_link = "No PDF Link"

        try:
            # Wait for the Graph link to be present and extract it
            graph_link = cells[1].find_element(By.XPATH, ".//a[contains(@href, '.jpg')]").get_attribute("href")
        except:
            graph_link = "No Graph Link"

        # Append the links
        row_data.append(pdf_link)
        row_data.append(graph_link)
        # Extract Location (City, State)
        tlr = cells[2].text if len(cells) > 2 else "N/A"
        rpc = cells[3].text if len(cells) > 3 else "N/A"
        row_data.append(tlr)
        row_data.append(rpc)

        # Extract Ranking Information (Score, Rank)
        go = cells[4].text if len(cells) > 4 else "N/A"
        oi = cells[5].text if len(cells) > 5 else "N/A"
        pr = cells[6].text if len(cells) > 6 else "N/A"
        row_data.append(go)
        row_data.append(oi)
        row_data.append(pr)
        # Extract Location (City, State)
        city = cells[7].text if len(cells) > 7 else "N/A"
        state = cells[8].text if len(cells) > 8 else "N/A"
        row_data.append(city)
        row_data.append(state)

        # Extract Ranking Information (Score, Rank)
        score = cells[9].text if len(cells) > 9 else "N/A"
        rank = cells[10].text if len(cells) > 10 else "N/A"
        row_data.append(score)
        row_data.append(rank)
        print(row_data)
        # Append row data to the table data list
    table_data.append(row_data)

# Define column headers for the DataFrame
columns = ["Institute Code", "Institute Name", "PDF Link", "Graph Link","TLR","RPC","GO","OI","PR", "City", "State", "Score", "Rank"]

# Convert the cleaned table data into a pandas DataFrame
df = pd.DataFrame(table_data, columns=columns)

# Print the DataFrame
print(df)
# Close the browser
driver.quit()
```
