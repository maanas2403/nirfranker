```python
import requests
from io import BytesIO
import numpy as np
response = requests.get(image_url)
if response.status_code == 200:
    image_data = response.content
else:
    raise Exception(f"Failed to download image from URL: {image_url}")

# Convert image data to numpy array for OpenCV
image_np = np.array(bytearray(image_data), dtype=np.uint8)
image = cv2.imdecode(image_np, cv2.IMREAD_COLOR)

# Check if the image was loaded correctly
if image is None:
    raise ValueError("Failed to load the image from the URL.")

# Perform OCR on the image
result = reader.readtext(image)

# Display the image using matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
for (bbox, text, prob) in result:
    print(f"Detected text: {text} (Confidence: {prob:.2f})")

# If you want to draw bounding boxes around detected text
for (bbox, text, prob) in result:
    top_left = tuple(map(int, bbox[0]))
    bottom_right = tuple(map(int, bbox[2]))
    image = cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)
```

```python
def process_image_and_extract(url):
    search_text = "Score"
    try:
        # Download the image
        response = requests.get(url)
        if response.status_code != 200:
            raise Exception(f"Failed to download image from URL: {url}")
        
        # Convert image data to numpy array for OpenCV
        image_np = np.array(bytearray(response.content), dtype=np.uint8)
        image = cv2.imdecode(image_np, cv2.IMREAD_COLOR)

        # Perform OCR on the image
        result = reader.readtext(image)

        # Check if there are enough results
        start_index = None
        for i, (_, text, _) in enumerate(result):
            if search_text.lower() in text.lower():
                start_index = i
                break
        
        if start_index is not None and len(result) > start_index + 1:
            # Extract values starting from the row after the found text
            start_index += 1  # Start after the found text
            end_index = start_index + 17  # We want up to 16 values (40th to 56th values)
            print([result[i][1] for i in range(start_index, min(end_index, len(result)))])
            return [result[i][1] for i in range(start_index, min(end_index, len(result)))]
        else:
            print("None")
            return [None] * 17  # Not enough results, return None for these columns
    except Exception as e:
        print(f"Error processing image from {url}: {e}")
        return [None] * 17
```

```python
df.iloc[:, -17:] = df['Graph Link'].apply(process_image_and_extract).tolist()
```
